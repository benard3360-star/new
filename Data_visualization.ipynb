# Data Visualization
## 1. Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

### (a). plot & line chart
for plot & line chart, we use the syntax:   
`plt.plot(x, y, 'line/marker style')`

plot with different markers

```python
# plt.plot(3, 2, ".")
# plt.plot(3, 2, "o")
# plt.plot(3, 2, "v")
# plt.plot(2, 2, "s")
# plt.plot(2, 2, "2")
# plt.plot(2, 2, "+")
# plt.plot(2, 2, "D")
plt.plot(2, 2, "h")
```

```python
# line chart
x = [1, 2, 3, 4, 5, 6]
y = x
# plt.plot(x, y)
# plt.plot(x, y, "--")
plt.plot(x, y, ":")
```

```python
# two line charts in one figure
# we can also specify line style combined with marker
x = np.array([1, 2, 3, 4, 5, 6])
y1 = x
y2 = x**2

plt.plot(x, y1, '-o')
plt.plot(x, y2, '-o')
```

### (b). scatter plot
for scatter plot, we use the syntax:   
`plt.scatter(x, y)`

```python
x = np.array([1,2,3,4,5,6,7,8])
y = x

plt.scatter(x, y)
```

### (c). bar chart
for bar chart, we use the syntax:   
`plt.bar(x, y)`

```python
x = np.array([1,2,3,4,5,6,7,8])
y = x

plt.bar(x, y)
```

### (d). markersize, color, width, label, legend, title
markersize, color, width:   
add arguments in the function

label, legend, title:   
`plt.xlabel(name)`, `plt.ylabel(name)`, `plt.legend( )`, `plt.title(name)`

```python
x = np.array([1,2,3,4,5,6,7,8])
y = x

colors = ['green', 'red', 'yellow', 'blue', 'purple', 'pink', 'orange', 'grey']
plt.scatter(x, y, c=colors, s=100)
```

```python
x = np.array([1, 2, 3, 4, 5, 6])
y1 = x
y2 = x**2

plt.plot(x, y1, '-o', color='green', markersize=5)
plt.plot(x, y2, '-o', color='red', markersize=8)
```

```python
x = np.array([1,2,3,4,5,6,7,8])
y = x

plt.bar(x, y, color='purple', width=0.3)
```

```python
x = np.array([1, 2, 3, 4, 5, 6])
y1 = x
y2 = x**2

plt.plot(x, y1, '-o', color='green', markersize=5, label='linear')
plt.plot(x, y2, '-o', color='red', markersize=5, label='square')

# add a label to the x axis
plt.xlabel('x value')
# add a label to the y axis
plt.ylabel('y value')
# add a title
plt.title('Linear and Square')
# add a legend (uses the labels from plt.scatter)
plt.legend()
```

## 2. Matplotlib Using Pandas

creating chart:
`df.plot.kind( )`

`kind` :
- `'line'` : line plot (default)
- `'bar'` : vertical bar plot
- `'barh'` : horizontal bar plot
- `'hist'` : histogram
- `'box'` : boxplot
- `'kde'` : Kernel Density Estimation plot
- `'density'` : same as 'kde'
- `'area'` : area plot
- `'pie'` : pie plot
- `'scatter'` : scatter plot
- `'hexbin'` : hexbin plot


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# We have a dataframe of the number of immigrants from India,
# China, and Kenya to Canada from 1980 to 1994.
mydata = {
    'year': ['1980', '1981', '1982', '1983', '1984', '1985', '1986', '1987', '1988', '1989', '1990', '1991', '1992', '1993', '1994'],
    'India': [8880, 8870, 8147, 7338, 5704, 4211, 7150, 10189, 11522, 10343, 12041, 13734, 13673, 21496, 18620],
    'China': [5123, 6682, 3308, 1863, 1527, 1816, 1960, 2643, 2758, 4323, 8076, 14255, 10846, 9817, 13128],
    'Kenya': [291, 296, 221, 201, 218, 245, 244, 628, 984, 1140, 869, 605, 529, 478, 414],
    }

immi_canada_df = pd.DataFrame(mydata)
immi_canada_df
```

```python
# default is line chart
immi_canada_df.plot()
```

```python
immi_canada_df.plot.bar()
```

```python
immi_canada_df.plot.bar(x='year', y=['China', 'India'])
```

```python
# We can add labels and titles as additional arguments
immi_canada_df.plot.bar(x='year', y=['China', 'India'], title='Number of Migrants from China and India (1980-1994)', ylabel='Number of Migrants')
```

```python
# We can also add color as additional argument
immi_canada_df.plot.bar(x='year', y=['China', 'India'], title='Number of Migrants from China and India (1980-1994)', ylabel='Number of Migrants', color={"China": "pink", "India": "lightblue"})
```

## 3. Seaborn

heatmap:   
`sns.heatmap(data=df)`

violinplot:  
`sns.violinplot(data=df, x=df["name"], y=df["name"])`

we can also use [catplot()](https://seaborn.pydata.org/generated/seaborn.catplot.html):  
`sns.catplot(data=df, x="name", y="name", hue="name", kind="chat type")`

`kind=` :
- `'point'`
- `'bar'`
- `'count'`
- `'box'`
- `'violin'`
- `'boxed'`
- `'strip'`
- `'swarm'`

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### (a). Heatmap

```python
# Read the file into a variable flight_data
flight_data = pd.read_csv("flight_delays.csv", index_col="Month")
flight_data
# each entry shows the average arrival delay (in minutes) for a different
# airline and month (all in year 2015)
```

```python
# Set the width and height of the figure
plt.figure(figsize=(14,7))

# Add title
plt.title("Average Arrival Delay for Each Airline, by Month")

# Heatmap showing average arrival delay for each airline by month
sns.heatmap(data=flight_data, annot=True)

# Add label for horizontal axis
plt.xlabel("Airline")
```

### (b). Violinplot

```python
df = sns.load_dataset("titanic")
sns.violinplot(x=df["age"])
```

```python
sns.violinplot(data=df, x="age", y="class")
```

```python
sns.violinplot(data=df, x="age", y="class", hue="alive")
```

### (c). catplot()

```python
sns.catplot(data=df, x="age", y="class", hue="alive", kind="violin")
```

```python
sns.catplot(data=df, x="age", y="class", hue="alive", kind="violin")

# add a title
plt.title('Titanic Passenger Age and Class Distribution by Survival')
# add a label to the x axis
plt.xlabel('Age')
# add a label to the y axis
plt.ylabel('Seat Class')
```
